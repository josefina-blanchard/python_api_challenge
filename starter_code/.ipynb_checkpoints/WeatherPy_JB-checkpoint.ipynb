```python
#Setting up modules and dependencies
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
from api_keys import api_key
from citipy import citipy
import openweathermapy as owm
import time
```

```python
#Setting the latitudes and longitudes
latitude = (-90, 90)
longitude = (-180, 180)

#Creating empty list for the coordinates
coordinates = []    
latitudes = np.random.uniform(low=latitude[0], high=latitude[1], size=1500)
#len(latitudes) #commenting out after verifying
longitudes = np.random.uniform(low=longitude[0], high=longitude[1], size=1500)
#len(longitudes) #comment out after verifying
#combining the two 
coordinates = zip(latitudes, longitudes)
coordinates
```

```python
#Creating empty list to grab the random coordinates combinations
cities = [] 
#looping thorugh the cities' coordinations
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    #if statement to grab only unique combinations for the coordinates     
        #adding those unique values to the list of cities
    if city not in cities:    
        cities.append(city)   
        
len(cities) 
```

```python
#cities   #comment out the cities ones listed
cities
```

```python
#Defining the parameter for the API call to the openweathermapy wrapper
params = {
    "APPID" : api_key,
    "units" : "metric"
}
```

```python
#Retrieving one single city to test the call and make sure the end points are coming through
test = []
results = owm.get_current("london", **params)
test.append({"City": results("name"),
            "Country Code": results("sys.country"),
            "Date": results("dt"),
            "Temperature": results("main.temp_max"),
            "Latitude": results("coord.lat"),
            "Wind Speed": results("wind.speed"),
            "Humidity": results("main.humidity"),
            "Cloudiness": results("clouds.all")})
test
```

```python
#build a series of scatter plots to showcase the following relationships:
#Temperature (F) vs. Latitude
#Humidity (%) vs. Latitude
#Cloudiness (%) vs. Latitude
#Wind Speed (mph) vs. Latitude

#setting the list to save the values of the weather by city
weather = []

for city in cities:
    try:
        results = owm.get_current(city, **params)
    except:
        continue
    else:
        print(f"Retrieving results for {city}")
        weather.append({"City": results("name"),
                        "Country Code": results("sys.country"),
                        "Dated": results("dt"),
                        "Temperature": results("main.temp_max"),
                        "Latitude": results("coord.lat"),
                        "Wind Speed": results("wind.speed"),
                        "Humidity": results("main.humidity"),
                        "Cloudiness": results("clouds.all")})

#weather  #commenting out after verifying
```

```python
#Create DataFrame
weather_df = pd.DataFrame(weather)
weather_df.tail()
```

```python
#Perform a weather check on each city using a series of successive API calls.
#Include a print log of each city as it'sbeing processed (with the city number and city name).
#export city data to a csv file
weather_df.to_csv("Weather_and_coordinates_by_cities.csv", header=True)
```

```python
#Create scatter plots for the different data points collected
city_lat = weather_df["Latitude"]
city_temp = weather_df["Temperature"]
city_humidity = weather_df["Humidity"]

temperature_plot = plt.scatter(city_lat, city_temp, s=city_humidity, marker="o", facecolors="blue", edgecolors="black", alpha=0.40)
plt.title("City temperature by latitude")
plt.xlabel("Latitud")
plt.ylabel("Temperature")
plt.grid(True)
plt.savefig("City temperature by latitude")
plt.show()
```

```python
#Create the plot for latitud vs humidity
city_lat = weather_df["Latitude"]
city_temp = weather_df["Temperature"]
city_humidity = weather_df["Humidity"]

humidity_plot = plt.scatter(city_lat, city_humidity, s=city_temp, marker="D", facecolors="red", edgecolors="black", alpha=0.40)
plt.title("City humidity by latitude")
plt.xlabel("Latitud")
plt.ylabel("Humidity")
plt.grid(True)
plt.ylim(0,100)
plt.savefig("City himidity by latitude")
plt.show()
```

```python
#Create the plot for latitude vs cloudiness
city_lat = weather_df["Latitude"]
city_temp = weather_df["Temperature"]
city_cloud = weather_df["Cloudiness"]

cloudiness_plot = plt.scatter(city_lat, city_cloud, s=city_temp, marker="X", facecolors="yellow", edgecolors="black", alpha=0.40)
plt.title("City cloudiness by latitude")
plt.xlabel("Latitud")
plt.ylabel("Cloudiness")
plt.grid(True)
plt.ylim(0,100)
plt.savefig("City cloidiness by latitud")
plt.show()
```
